## MSV - Esercizio 1.1

```python
Image('Immagini/MSV_1.1_Testo.png',width=900)
```

__Lettura del testo__

* è un moto rettilineo uniformemente accelerato (l'accelerazione $a$ è costante).
* conosciamo le velocità $v_1$ e $v_2$ in due istanti temporali $t_1$ e $t_2$ e lo spazio percorso $\Delta x$, 
   dal punto materiale nell'intervallo $\Delta t = t_2-t_1$.
* dobbiamo calcolare $a$ e $\Delta t$.

__Metodo 1__

Usiamo la formula generale della legge oraria del punto materiale con accelerazione costante

$$
v(t) = v_0 + a(t-t_0) \\
x(t) = x_0+v_0(t-t_0)+\frac{1}{2}a(t-t_0)^2
$$

Dalla prima espriamiamo $a$ in funzione di $\Delta v$ e $\Delta t$
$$
v(t_2) = v_1 + a\Delta t \,\, \rightarrow \,\, a = \frac{\Delta v}{\Delta t}
$$
che inserita nella seconda (per $t=t_2$) da
$$
x(t_2) = x_1 + v_1\Delta t + \frac{1}{2}\Delta v \Delta t \,\, \rightarrow \,\,
\Delta x= \Delta t(v_1+\frac{1}{2}\Delta v) = \frac{\Delta t}{2}(v_1+v_2)
$$
Da queste relazioni calcolo $\Delta t$ e poi $a$
$$
\Delta t = \frac{12,5 m}{7,5 m/s} = 1.67 s \,\, e \,\,
a = \frac{5 m/s}{1,67 s}= 3 m/s^2
$$

__Metodo 2__

Esprimiamo l'accelerazione in termini della varaizione di $v$ rispetto alla _posizione_
$$
a = \frac{dv}{dt} = \frac{dv}{dx}\frac{dx}{dt} = v\frac{dv}{dx}
$$
Integriamo per quadrature sapendo che $a$ è costante
$$
a(x_2-x_1) = \frac{1}{2}(v_2^2-v_1^2) \,\, \rightarrow \,\,
a = \frac{v_2^2-v_1^2}{2\Delta x}
$$
Possiamo calcolare $a$ direttamente da questa formula e poi sostituire il suo valore nella $v(t)$ per il calcolo di $\Delta t$.

__Svolgimento punto (a)__

La velocità è costante quindi il percorso che minimizza il tempo complessivo è il più breve.

Dobbiamo minimizzare la lunghezza $L$ del percorso $AO+OB$. Usiamo il teorema di Pitagora
$$
L^2 = \big(x^2+h^2\big) + \big( (d-x)^2+h^2 \big) 
$$
Possiamo minimizzare diretttamente $L^2$ per non dover trattare le radici quadrate, così abbiamo
$$
2x-2(d-x)=0 \,\, \rightarrow \,\, x = \frac{d}{2}
$$
Il cammino più corto è quello per il quale $AO=OB$, ovvero per il quale l'angolo fra $AO$ e l'asse $x$
è uguale a quello fra $OB$ e l'asse $x$.

__Svolgimento punto (b)__

In questo caso la velocità è differente durante i 2 tratti quindi minimizziamo la somma dei tempi sui due
segmenti.

Usando il teorema del seno possiamo esprimere i seni degli angoli $\theta_1$ e $\theta_2$
$$
\frac{x}{sin(\theta_1)} = AO \,\, \rightarrow \,\, sin(\theta_1) = \frac{x}{\sqrt{x^2+h_1^2}} \\
\frac{d-x}{sin(\theta_2)} = OB \,\, \rightarrow \,\, sin(\theta_2) = \frac{d-x}{\sqrt{(d-x)^2+h_2^2}}
$$
Il tempo totale si scrive come
$$
t = \frac{\sqrt{x^2+h_1^2}}{v_1} + \frac{\sqrt{(d-x)^2+h_2^2}}{v_2}
$$
e ne cerchiamo il minimo annullando la derivata prima rispetto a $x$
$$
%\frac{dt}{dx} = \frac{sin(\theta_1)}{v_1} - \frac{sin(\theta_2)}{v_2}  = 0 
\frac{dt}{dx} = \frac{1}{v_1}\frac{x}{\sqrt{x^2+h_1^2}} -\frac{1}{v_2} \frac{d-x}{\sqrt{(d-x)^2+h_2^2}}  = 0
$$
che è proprio la condizione 
$$
 \frac{sin(\theta_1)}{v_1} = \frac{sin(\theta_2)}{v_2}
$$

__Lettura del testo__

* è un moto rettilineo vario (accelerazione non costante).
* conosciamo posizione e velocita a $t=0$.


__Svolgimento__

Integriamo l'equazione per la velocità (espressa in funzione del tempo)
$$
a = \frac{dv}{dt} = -k v^2
$$
da risolvere insieme alla condizione iniziale $v(0)=v_0$. Si risolve per separazione di variabili, ovvero
separando nei due membri i differenziali insieme alle relative variabili
$$
\frac{dv}{v^2} = -k dt \, \rightarrow \, 
\int_{v_0}^{v} \frac{dv}{v^2} = -k \int_{0}^{t}dt 
$$
da cui otteniamo
$$
-\left(\frac{1}{v}-\frac{1}{v_0}\right) = -kt \, \rightarrow \,
v (t)= \frac{v_0}{1+kv_0 t}
$$
Usare l'analisi dimensionale per controllare il risultato! Quali sono le dimensioni fisiche di k?
Studiate l'andamento di $v$ in funzione del tempo (per $k$ positivo).

Ripetiamo l'analsi in funzione dello spazio
$$
a = \frac{dv}{dt} = \frac{dv}{dx}  \frac{dx}{dt} =  \frac{dv}{dx} v = -k v^2
$$
e quindi nello stesso modo di prima
$$
\int_{v_0}^{v} \frac{dv}{v} = -k \int_{0}^{x}dx
$$
da cui otteniamo
$$
ln\big(\frac{v(x)}{v_0}\big) = -kx \, \rightarrow \,
v(x) = v_0 e^{-kx}
$$

Infine, per determinare la legge oraria possiamo integrare la $v(t)$, oppure uguagliamo le due espressioni per la velocità
$$
v(x(t)) = v_0e^{-k(x(t))} = \frac{v_0}{1+kv_0 t} \, \rightarrow \, 
x(t) = -\frac{1}{k}ln\left(\frac{1}{1+kv_0 t}\right) = 
\frac{1}{k}ln\left(1+kv_0 t\right)
$$

## MSV - Esercizio 1.17

```python
Image('Immagini/MSV_1.17_Testo.png',width=900)
```

__Lettura del testo__

* è un moto nel piano. Si scompone in due moti rettilinei _indipendenti_ lungo gli assi $x$ e $y$.
  L'accelerazione (dovuta alla forza di gravità) è diretta lungo $y$.
* conosciamo posizione e modulo della velocita a $t=0$.

__Svolgimento__

_Moto lungo l'asse $x$_

E' un moto uniforme ($a_x=0$) con velocità pari alla velocità iniziale $v_x = v cos(\alpha)$
$$
x(t) = v cos(\alpha) t
$$
Al tempo $t=t_f$ la componente $x$ del moto raggiunge la distanza orizzontale 
$D_x= \sqrt{D^2-h^2}$  fra il cannone e il bersaglio
$$
v cos(\alpha) t_f = D_x \, \, \quad
t_f = \frac{D_x}{v cos(\alpha)}
$$
Dall'analisi del moto lungo $x$ abbiamo determinato il valore del tempo a cui deve avvenire l'impatto. 

Ora studiamo il moto lungo $y$ e richiediamo che per $t=t_f$ la compoenente $y$ della posizione sia uguale a $h$.

_Moto lungo l'asse $y$_

E' un moto uniformente accelerato ($a_y=-g$) con velocità inziale $v_y = v sin(\alpha)$
$$
y(t) = v sin(\alpha) t - \frac{1}{2}g t^2
$$
Ora imponiamo che $y(t_f)=h$
$$
v sin(\alpha)  \frac{D_x}{v cos(\alpha)} - \frac{1}{2}g  \frac{D_x^2}{v^2 cos^2(\alpha)} = h
$$
che si può riscrivere come una equazione di secondo grado in $tg(\alpha)$
$$
D_x tg(\alpha) - \frac{gD_x^2}{2v^2}\left(1+tg^2(\alpha)\right) = h
$$
che riscriviamo in forma "canonica" $X^2 + bX +c = 0$
$$
X^2 -\frac{2v^2}{g D_x} X + \frac{2hv^2}{gD_x^2}+ 1 = 0
$$

Abbiamo due soluzioni possibili
$$
X_{1,2} = -\frac{v^2}{g D_x} \pm \sqrt{\frac{v^4}{g^2 D^2_x} - \frac{2hv^2}{gD_x^2}- 1 }
$$
e quindi ci sono a seconda del valore del discriminante $\Delta$ (argomento della radice) i seguenti casi __fisicamente__
ammissibili:
* due soluzioni distinte per $\Delta > 0$
* una soluzione (ovvero due soluzioni coincidenti) per $\Delta=0$
* nessuna soluzione per $\Delta < 0$

Nel caso del testo abbiamo

```python
g = 9.8
v = 3e2
D,h = 5e3,1e3
```

```python
Dx = np.sqrt(D**2-h**2)
b = -v**2/(g*Dx)
Delta = np.sqrt(v**4/(g**2*Dx**2)-2*h*v**2/(g*Dx**2)-1)
```

```python
X1,X2 = -b+Delta,-b-Delta
X1,X2
```

E calcoliamo il valore dell'angolo in gradi

```python
180/np.pi*np.arctan(X1),180/np.pi*np.arctan(X2)
```

## MSV - Esercizio 1.15

__Lettura del testo__

* il punto $P$ è vincolato a muoversi lungo l'asse $x$ e la biella è un vincolo rigido che lega le posizioni di $A$ e di $P$.
* $A$ descrive un moto circolare uniforme sulla circonferenza di raggio $R$.
* dobbiamo esprimere la posizione di $P$ in funzione di quella di $A$, ovvero in funzione di $\theta$. Nota la P(t) calcoliamo la sua velocità e accelerazione
   per derivazione.

__Svolgimento__

Usando il teorema dei seni abbiamo che
$$
sin(\theta') = \frac{R}{b}sin(\theta) \,\, \rightarrow \,\, 
cos(\theta') = \sqrt{1-\frac{R^2}{b^2}sin^2(\theta)}
$$
e possiamo determinare la posizione di $P$ in funzione di $\theta$
$$
x_P(t) = R cos(\theta) +  bcos(\theta') = R cos(\theta) + \sqrt{b^2-R^2sin^2(\theta)}
$$
valida per $b \geq R$, visto che $P$ è vincolato sull'asse $x$.

Infine usiamo la condizione di velocità angolare costante per espimere $\theta=\omega t$ e abbiamo
la _legge oraria_ del moto di $P$
$$
x_P(t) = R cos(\omega t) + b\sqrt{1-\frac{R^2}{b^2}sin^2(\omega t)}
$$
Abbiamo messo $b$ a fattore nel secondo addendo, questo aiuta la lettura dell'equazione e permette
considerazioni qualitative sul moto. 

Se $b \gg R$ la legge oraria si riduce a 
$$
x_P(t) = R cos(\omega t) + b
$$
che ci dice che in questo limite la $P$ descrive un moto armonico di pulsazione $\omega$ (e in fase con $A$) ma
traslato attorno al punto $x=b$.

Velocità e accelerazione si determinano derivando la $x_P(t)$ rispetto a $t$
$$
v_P(t) = -\omega R sin(\omega t) - \frac{\omega R^2}{b} \frac{sin(\omega t) cos(\omega t)}{\sqrt{1-\frac{R^2}{b^2}sin^2(\omega t)}} = 
-\omega R sin(\omega t) - \frac{\omega R^2}{2b} \frac{sin(2\omega t) }{\sqrt{1-\frac{R^2}{b^2}sin^2(\omega t)}}
$$

$$
a_P(t) = -\omega^2 R cos(\omega t) - \frac{\omega R^2}{2b} \Big( \frac{2\omega cos(2\omega t)}{\sqrt{1-\frac{R^2}{b^2}sin^2(\omega t)}}
-\frac{\omega R^2}{b^2} \frac{sin^2(2\omega t)}{\big(1-\frac{R^2}{b^2}sin^2(\omega t)\big)^{3/2}}
\Big)
$$
